```python
classifier = pipeline(
    task="zero-shot-audio-classification", model="laion/clap-htsat-unfused"
)
```

```python
sounds = {"Sound of a trumpet" : 4,
"Sound of a guitar" : 1,
"Sound of only a piano" : 3,
"Sound of a violin" : 5,
"Tonal Harmony" : 2,
"Electronic Beats" : 0, 
}
```

```python
import csv
csv_file_path = 'C:/Users/bruno/OneDrive - Imperial College London/Pers/Portfolio/Sound Classifier/nuwehack-data-AI_SR/data/labels_paths_train.csv'

idxs = []
labels = []
paths = []

with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)  
    for row in csvreader:
        idx, label, path = row
        idxs.append(idx)
        labels.append(label)
        paths.append(path)

print("Idxs:", idxs)
print("Labels:", labels)
print("Paths:", paths)
```

```python
candidate_labels = ["Sound of a trumpet", "Sound of a guitar", "Sound of only a piano", "Sound of a violin", "Tonal Harmony", "Electronic Beats"]
```

```python
import librosa
sample_rate = 22050
audio_file_path = "C:/Users/bruno/OneDrive - Imperial College London/Pers/Portfolio/Sound Classifier/nuwehack-data-AI_SR/data/train/audio_2.wav"
audio_data, sample_rate = librosa.load(audio_file_path, sr=None)
res = classifier(audio_data, candidate_labels=candidate_labels)
print(res)
```

```python
import librosa

predictions = []
pred = []
i = 0
local_path = "C:/Users/bruno/OneDrive - Imperial College London/Pers/Portfolio/Sound Classifier/nuwehack-data-AI_SR/data/"
for path in paths:

    audio_file_path = local_path + path  
    audio_data, sr = librosa.load(audio_file_path, sr=None)
    prediction = classifier(audio_data, candidate_labels=candidate_labels)
    predictions.append(prediction[0]["label"])
    ##print(prediction[0]["label"], " " , path, " ", prediction[0]["score"])
    if sounds[prediction[0]["label"]] == int(labels[i]):
        pred.append(1)
    else :
        pred.append(0)
        print("Pred ", prediction[0]["label"], "was ", int(labels[i]))
        print(prediction[0]["label"], " " , path, " ", prediction[0]["score"])
    i += 1
    average_value = sum(pred) / len(pred)
    print("The average is:", average_value)
print(predictions)
```
