```python
def calculate_percentage(lst):
    from collections import Counter
    count = Counter(lst)  # Count the occurrence of each element
    total = len(lst)  # Total number of elements
    percentages = {element: (count[element] / total) * 100 for element in count}
    return percentages

```

```python
import json

def save_predictions_to_json(labels, predictions):
    # Ensure the labels and predictions lists are of the same length
    if len(labels) != len(predictions):
        raise ValueError("Labels and predictions lists must be of the same length.")

    # Create the dictionary structure as specified
    predictions_dict = {"target": dict(zip(labels, predictions))}

    # Write the dictionary to predictions.json
    with open('predictions.json', 'w') as file:
        json.dump(predictions_dict, file, indent=4)

print(len(idxs_test))
save_predictions_to_json(idxs_test, pred_num)

       
```
